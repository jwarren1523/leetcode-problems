# Idea
convert into a real number, sum, then convert back to list node? 
Or go through node by node and carry over "10" if needed. (can't carry over more than 10)
Need to link node back

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def add_two_numbers(l1: ListNode, l2:ListNode) -> ListNode:
    sum = []
    carry_over = 0
    first_run = True
    previous_node = None
    a=l1
    b=l2
    while a or b or carry_over:
        a_val = a.val if a else 0
        b_val = b.val if b else 0
        node_sum = a_val+b_val+carry_over
        carry_over = 1 if node_sum >= 10 else 0
        node = ListNode(node_sum % 10)
        sum.append(node)
        if not first_run:
            previous_node.next = node
        else: 
            first_run = False
        previous_node = node
        a=a.next if a else None
        b=b.next if b else None

    return sum[0]


```

# Evaluation

Don't need to specify a != None, just can simply use "a" as a boolean

```python
add_two_numbers_test = add_two_numbers

# Helper to convert linked list â†’ Python list
def listnode_to_list(node):
    result = []
    loop_node = node
    while loop_node:
        result.append(loop_node.val)
        loop_node = loop_node.next
    return result
    
def listnode_from_list(lst):
    if not lst:
        return None
    head = ListNode(lst[0])
    current = head
    for val in lst[1:]:
        current.next = ListNode(val)
        current = current.next
    return head

# Test 0: Simple example
l1 = listnode_from_list([2,4,3])
l2 = listnode_from_list([5,6,4])
assert listnode_to_list(add_two_numbers_test(l1, l2)) == [7,0,8]
print("Test 0 passes")


# Test 1: Both zeros
l1 = listnode_from_list([0])
l2 = listnode_from_list([0])
assert listnode_to_list(add_two_numbers_test(l1, l2)) == [0]
print("Test 1 passes")


# Test 2: Different lengths
l1 = listnode_from_list([1,2,3])
l2 = listnode_from_list([4,5])
assert listnode_to_list(add_two_numbers_test(l1, l2)) == [5,7,3]
print("Test 2 passes")

# Test 3: Different lengths and carry over
l1 = listnode_from_list([9,9,9,9,9,9,9])
l2 = listnode_from_list([9,9,9,9])
assert listnode_to_list(add_two_numbers_test(l1, l2)) == [8,9,9,9,0,0,0,1]
print("Test 3 passes")
```
