# My solution

```python
class Solution: 
    def calculate_container_size(self, height:list[int]) -> int:
        top_score = 0
        for i in range(len(height)-1):
            for j in range(i+1, len(height)):
                water_height = min(height[i], height[j])
                distance = j-i
                water_area = water_height * distance
                if water_area > top_score:
                    top_score = water_area
        return top_score
```

# Chat GPT's solution

```python
class Solution_a:
    def calculate_container_size(self, height: list[int]) -> int:
        left, right = 0, len(height) - 1
        max_area = 0
        while left < right:
            water_height = min(height[left], height[right])
            distance = right - left
            max_area = max(max_area, water_height * distance)
            # Move the smaller side inwards
            if height[left] < height[right]:
                left += 1
            else:
                right -= 1
        return max_area
```

# Evaluation

This:
    `if water_area > top_score:`
        `top_score = water_area`
can be replaced by this:
    `top_score = max(water_area, top_score)`

ChatGPT's solution starts 2 pointers at the greatest distance and then whatever side has the lowest value, moves inwards. This works because you use the lower of the two values. 

# Tests

```python
test_solution = Solution_a()
test_function = test_solution.calculate_container_size

def test(inputs:dict, answer: str, t_number:int) -> None:
    print("---------------------")
    print(f"--Test-{t_number}--")
    output = test_function(**inputs)
    print(f"expected: {answer}")
    print(f"actual:   {output}")
    assert output == answer
    print(f"Test {t_number} passes")
    print("---------------------")



print("---------------------")
# -----------------------
# From Prompt
# -----------------------

# Test 1 — prompt example
test({"height": [1,8,6,2,5,4,8,3,7]},
     49, 1)

# Test 2 — prompt example
test({"height": [1,1]},
     1, 2)

# -----------------------
# Additional Tests
# -----------------------

# Test 3 — increasing heights
test({"height": [1,2,3,4,5]},
     6, 3)   # container between 1 and 5 → min(1,5)*4 = 4, but better is 2 & 5 → 2*3 = 6

# Test 4 — decreasing heights
test({"height": [5,4,3,2,1]},
     6, 4)

# Test 5 — all zeros
test({"height": [0,0,0,0]},
     0, 5)

# Test 6 — large plateaus
test({"height": [4,4,4,4]},
     12, 6)  # min(4,4)*3 = 12

# Test 7 — one tall, many small
test({"height": [10,1,1,1,10]},
     40, 7)  # min(10,10)*4

# Test 8 — alternating highs and lows
test({"height": [2,9,2,9,2]},
     18, 8)  # min(9,9)*2

# Test 9 — minimal valid length
test({"height": [0,10]},
     0, 9)

# Test 10 — random mid-sized
test({"height": [3,1,2,4,5,2,6]},
     18, 10)  # between 3 (i=0) and 6 (i=6): min(3,6)*6 = 18

print("---------------------")
print("All tests pass!")
print("---------------------")
```
