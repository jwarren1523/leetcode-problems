# Idea
## When 3 rows
separate indexes into groups of 4
1 goes on first line, 2 and 4 go on middle line, 3 goes on last line. Then combine together

## broadly
rows 1 -> cycle of 1
rows 2 -> cycle of 2: 1|2
rows 3 -> cycle of 4: 1|2,4|3
rows 4 -> cycle of 6: 1|2,6|3,5|4

from rows 2, cycle = 2 + (rows-2)*2

using indexes:
    if index divisble by cycle -> row 1
    if index is divisble by rows -> last row (row number)
    else
        if index is <= row number -> row index+1 (actually the same for first 2 ifs)
        if index is > row number -> row (cycle-index) +1

```python
class Solution: 
    def convert_to_zigzag(self, s:str, num_rows:int) -> str:
        if num_rows == 1:
            return s
        cycle = 2 + (num_rows-2)*2
        rows = {i:[] for i in range(num_rows)}
        for i in range(len(s)):
            i_mod = i % cycle
            if i_mod <= num_rows-1:
                rows[i_mod].append(s[i])
            else:
                row = cycle - i_mod
                rows[row].append(s[i])
        output = ""
        for j in rows:
            complete_row = "".join(rows[j])
            output += complete_row
        return output
```

# Evaluation

To join a string from lists: `"".join(*list*)` where "" is the delimiter (empty in this case)

This is called a generator `rows = {i:[] for i in range(num_rows)}`
I could have done the generator as so: `rows = [[] for _ in range(num_rows)]` 

```python
solution = Solution()

def test(inputs:dict, answer: str, t_number:int) -> None:
    print("---------------------")
    print(f"--Test-{t_number}--")
    output = solution.convert_to_zigzag(**inputs)
    print(f"expected: {answer}")
    print(f"actual:   {output}")
    assert output == answer
    print(f"Test {t_number} passes")
    print("---------------------")


print("---------------------")
# -----------------------
# From Prompt
# -----------------------

# Test 1
test({"s": "PAYPALISHIRING", "num_rows": 3},
     "PAHNAPLSIIGYIR", 1)

# Test 2
test({"s": "PAYPALISHIRING", "num_rows": 4},
     "PINALSIGYAHRPI", 2)

# Test 3
test({"s": "A", "num_rows": 1},
     "A", 3)

# -----------------------
# Additional Edge Cases
# -----------------------

# Test 4 — single character and many rows
test({"s": "X", "num_rows": 5},
     "X", 4)

# Test 5 — rows >= length
test({"s": "HELLO", "num_rows": 10},
     "HELLO", 5)

# Test 6 — 2-row simple zigzag
test({"s": "ABCDE", "num_rows": 2},
     "ACEBD", 6)

# Test 7 — multi-row irregular pattern
test({"s": "THISISAZIGZAG", "num_rows": 3},
     "TIIGHSSZGAIAZ", 7)

# Test 8 — punctuation and mixed chars
test({"s": "a,b.c,d.e", "num_rows": 3},
     "ace,.,.bd", 8)
print("---------------------")
print("All tests pass!")
print("---------------------")
```
