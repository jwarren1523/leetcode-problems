# Idea
Already got a solution, but need to work out how to avoid overflow



# My solution

```python
class Solution: 
    def reverse_integer(self, num:int) -> int:
        overflow = False
        INT_MIN = -2**31
        INT_MAX =  2**31 - 1
        if overflow:
            return 0
        is_negative = num < 0
        num_list = [c for c in str(num)]
        if is_negative: num_list = num_list[1:]
        
        num_list = num_list[::-1]
        reverse_num = int("".join(num_list))
        if reverse_num < INT_MIN or reverse_num > INT_MAX:
            return 0
        if is_negative:
            reverse_num = -reverse_num
        
        return reverse_num
```

# Evaluation

I don't need a list, I can reverse strings with `[::-1]`

When having a constraint of 32 bit, technically turning into a string a reversing is cheating. Numbers don't overflow in Python (unlike C), so it's actually okay, but simulating a 32 bit environment, this method would not work.
See other chatGPT solution. 

# Chat GPT's solution

```python
# String based solution:

class Solution_a1:
    def reverse_integer(self, x: int) -> int:
        INT_MIN = -2**31
        INT_MAX =  2**31 - 1

        sign = -1 if x < 0 else 1
        rev = int(str(abs(x))[::-1])

        rev *= sign

        if rev < INT_MIN or rev > INT_MAX:
            return 0

        return rev
```

```python
# Non-string based solution

class Solution_a2:
    def reverse_integer(self, x: int) -> int:
        INT_MIN = -2**31        # -2147483648
        INT_MAX =  2**31 - 1    # 2147483647

        result = 0
        sign = -1 if x < 0 else 1
        x = abs(x)

        while x != 0:
            digit = x % 10
            x //= 10

            # Overflow check before multiplication/addition
            if (result > INT_MAX // 10 or 
               (result == INT_MAX // 10 and digit > 7)):
                return 0

            result = result * 10 + digit

        return sign * result
```

# Tests

```python
test_solution = Solution()
test_function = test_solution.reverse_integer

def test(inputs:dict, answer: int, t_number:int) -> None:
    print("---------------------")
    print(f"--Test-{t_number}--")
    output = test_function(**inputs)
    print(f"expected: {answer}")
    print(f"actual:   {output}")
    assert output == answer
    print(f"Test {t_number} passes")
    print("---------------------")


print("---------------------")
# -----------------------
# From Prompt
# -----------------------

# Test 1 — example 1
test({"num": 123},
     321, 1)

# Test 2 — example 2
test({"num": -123},
     -321, 2)

# Test 3 — example 3
test({"num": 120},
     21, 3)

# -----------------------
# Additional Edge Cases
# -----------------------

# Test 4 — zero
test({"num": 0},
     0, 4)

# Test 5 — single digit positive
test({"num": 7},
     7, 5)

# Test 6 — single digit negative
test({"num": -8},
     -8, 6)

# Test 7 — overflow positive (should return 0)
# reversed: 8463847412 → overflow
test({"num": 2147483648},
     0, 7)

# Test 8 — overflow negative (should return 0)
# reversed: -9463847412 → overflow
test({"num": -2147483649},
     0, 8)

# Test 9 — large valid input
test({"num": 1534236469}, 0, 9)  # known to overflow when reversed

# Test 10 — number ending in many zeros
test({"num": 100000},
     1, 10)

print("---------------------")
print("All tests pass!")
print("---------------------")
```
