# Plan

group into number blocks in array, then count at the end

```python
special_pairs = {"IV", "IX", "XL", "XC", "CD", "CM"}

def numeral_converter_1 (numeral:str) -> int:
    count = 0
    for i in range(len(numeral)):
        char = numeral[i]
        char_value = char_converter(char)
        if i < len(numeral)-1:
            char2 = numeral[i+1]
            combine_char = char+char2
            char_should_be_negative = combine_char in special_pairs
            if char_should_be_negative:
                char_value = -char_value
        count += char_value
    return count

def char_converter (char:str) -> int:
    if len(char) > 1:
        raise Exception
    match char:
        case "I":
            return 1
        case "V":
            return 5
        case "X":
            return 10
        case "L":
            return 50
        case "C":
            return 100
        case "D":
            return 500
        case "M":
            return 1000
```

```python
# ChatGPT solution:

def numeral_converter_a1(s: str) -> int:
    roman_map = {'I': 1, 'V': 5, 'X': 10, 'L': 50,
                 'C': 100, 'D': 500, 'M': 1000}
    total = 0
    prev_value = 0
    for char in reversed(s):
        value = roman_map[char]
        if value < prev_value:
            total -= value
        else:
            total += value
        prev_value = value
    return total
```

# Evaluation

Maps are better than switch statements

chatGPT's solution goes in reverse then sees if the next character has a lower value than the previous

```python
numeral_converter_test_function = numeral_converter_a1

# Test 0:
assert numeral_converter_test_function("") == 0
print ("Test 0 passes")

# Test 1: ordinary
assert numeral_converter_test_function("III") == 3
print ("Test 1 passes")

# Test 2: larger numbers
assert numeral_converter_test_function("LVIII") == 58
print ("Test 2 passes")

# Test 3: with subtraction
assert numeral_converter_test_function("MCMXCIV") == 1994
print ("Test 3 passes")

# Test 4: edge cases
assert numeral_converter_test_function("CCCCCCCCCCCCCCC") == 1500
print ("Test 4 passes")
```
