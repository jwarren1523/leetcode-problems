# Plan

group into number blocks in array, then count at the end

```python
special_pairs = {"IV", "IX", "XL", "XC", "CD", "CM"}

def numeral_converter_1 (numeral:str) -> int:
    count = 0
    for i in range(len(numeral)):
        char = numeral[i]
        char_value = char_converter(char)
        if i < len(numeral)-1:
            char2 = numeral[i+1]
            combine_char = char+char2
            char_should_be_negative = combine_char in special_pairs
            if char_should_be_negative:
                char_value = -char_value
        count += char_value
    return count

def char_converter (char:str) -> int:
    if len(char) > 1:
        raise Exception
    match char:
        case "I":
            return 1
        case "V":
            return 5
        case "X":
            return 10
        case "L":
            return 50
        case "C":
            return 100
        case "D":
            return 500
        case "M":
            return 1000
```

```python
numeral_converter_test_function = numeral_converter_1

# Test 1: ordinary
assert numeral_converter_test_function("III") == 3
print ("Test 1 passes")

# Test 2: larger numbers
assert numeral_converter_test_function("LVIII") == 58
print ("Test 2 passes")

# Test 3: with subtraction
assert numeral_converter_test_function("MCMXCIV") == 1994
print ("Test 3 passes")
```
