# Idea
go through each letter of first word
see if there is any word in that list without that letter

find min word length first?

```python
def common_prefix_calculator (strings : list[str]) -> str:
    reference_word = strings[0]
    for letter_idx in range(len(reference_word)):
        # print('------------')
        # print(f"reference_word[letter_idx]: {reference_word[letter_idx]}")
        for word in strings:
            if len(word) < letter_idx+1:
                return reference_word[:letter_idx]
            # print(f"word[letter_idx]: {word[letter_idx]}")
            if word[letter_idx] != reference_word[letter_idx]:
                # print(f"letter_idx: {letter_idx}")
                return reference_word[:letter_idx]
    return reference_word
```

```python
# ChatGPT's solution

def common_prefix_calculator(strs: list[str]) -> str:
    if not strs:
        return ""
    strs.sort()
    first, last = strs[0], strs[-1]
    i = 0
    while i < len(first) and i < len(last) and first[i] == last[i]:
        i += 1
    return first[:i]
```

# Evaluation

To avoid coverin the same word again - strings[1:]

GPT's solution sorts the strings and compares the first and last - after sorting, middle ones don't matter. Elegant but not necessarily the most efficient. This while loop is cleaner too. 

```python
common_prefix_calculator_test  = common_prefix_calculator  

def test(strings: str, answer: str) -> None:
    output = common_prefix_calculator_test(strings)
    assert output == answer

# Test 0: empty tests
test(["flower","bow","flight"], "")
test(["flower","","flight"], "")
test(["","bow","flight"], "")
test(["","",""], "")
print (f"Test 0 passes")

# Test 1: Ordinary tests
strings = ["flower","flow","flight"]
answer = "fl"
test(strings, answer)
print (f"Test 1 passes")

# Test 2: No similarities
strings = ["dog","racecar","car"]
answer = ""
test(strings, answer)
print (f"Test 2 passes")

# Test 3: Same words
test(["dog","dog","dog"], "dog")
test(["dog"], "dog")
test(["dog","doge","doge"], "dog")
print (f"Test 3 passes")

# Test 4: boundary
test(["a"*200, "b"*200], "")
test(["a"] * 200 , "a")
print (f"Test 4 passes")
```
