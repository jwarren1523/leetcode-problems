```python
def common_prefix_calculator (strings : list[str]) -> str:
    return strings[0]
```

```python
common_prefix_calculator_test  = common_prefix_calculator  

# Test 1:
strings = ["flower","flow","flight"]
assert common_prefix_calculator_test(strings) == "fl"
print ("Test 0 passes")

# Test 2:
strings = ["dog","racecar","car"]
assert common_prefix_calculator_test(strings) == ""
print ("Test 0 passes")
```
