# Idea

Have a queue
move along string, 
    everytime a new opening bracket, add to the queue, 
    everytime closing bracket, if it matches the latest item in the queue, remove from queue
    otherwise throw an error

```python
def is_valid(brackets:str) -> bool:
    opening_bracket = {'(','[','{'}
    closing_bracket = {'}',']',')'}
    matching_bracket = {")":"(", "]":"[", "}":"{", }
    queue = []
    if brackets[0] in closing_bracket: return False
    if len(brackets) == 1: return False

    for b in brackets:
        if b in opening_bracket:
            queue.append(b)
        if b in closing_bracket:
            if matching_bracket.get(b) != queue[-1]:
                return False
            else:
                queue.pop()
    if queue:
        return False
    return True
```

```python
# ChatGPT's Solution:

def is_valid_a(s: str) -> bool:
    stack = []
    pairs = {')': '(', ']': '[', '}': '{'}

    for char in s:
        if char in pairs.values():         # opening bracket
            stack.append(char)
        else:                               # closing bracket
            if not stack or stack[-1] != pairs[char]:
                return False
            stack.pop()

    return len(stack) == 0
```

# Evaluation
Using a queue was a good idea -> though technically it was a stack! 

Using sets for keeping brackets was a good idea, but actually could have just used matching_brackets.values()

I don't need to do separate return true or false at the end - just one simple boolean return would suffice

```python
is_valid_test  = is_valid_a

# Test 0: single bracket
assert is_valid_test("(") == False
print (f"Test 0 passes")

# Test 1: simple test
assert is_valid_test("()") == True
print (f"Test 1 passes")

# Test 2: all brackets
assert is_valid_test("()[]{}") == True
print (f"Test 2 passes")

# Test 3: non-matching brackets
assert is_valid_test("(]") == False
print (f"Test 3 passes")

# Test 4: nested brackets
assert is_valid_test("([])") == True
print (f"Test 4 passes")

# Test 5: incorrectly nested brackets
assert is_valid_test("([)]") == False
print (f"Test 5 passes")

# Test 6: first bracket closes
assert is_valid_test("]()]") == False
print (f"Test 6 passes")

# Test 7: long sequence but missing one close
assert is_valid_test("(((((((((()))))))))") == False
print (f"Test 7 passes")

```
