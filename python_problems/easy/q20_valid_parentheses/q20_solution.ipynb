# Idea

Have a queue
move along string, 
    everytime a new opening bracket, add to the queue, 
    everytime closing bracket, if it matches the latest item in the queue, remove from queue
    otherwise throw an error

```python
def is_valid(brackets:str) -> bool:
    opening_bracket = {'(','[','{'}
    closing_bracket = {'}',']',')'}
    matching_bracket = {")":"(", "]":"[", "}":"{", }
    queue = []
    if brackets[0] in closing_bracket: return False
    if len(brackets) == 1: return False

    for b in brackets:
        if b in opening_bracket:
            queue.append(b)
        if b in closing_bracket:
            if matching_bracket.get(b) != queue[-1]:
                return False
            else:
                queue.pop()
    return True
```

```python
is_valid_test  = is_valid  

# Test 0
assert is_valid_test("(") == False
print (f"Test 0 passes")

# Test 1
assert is_valid_test("()") == True
print (f"Test 1 passes")

# Test 2
assert is_valid_test("()[]{}") == True
print (f"Test 2 passes")

# Test 3
assert is_valid_test("(]") == False
print (f"Test 3 passes")

# Test 4
assert is_valid_test("([])") == True
print (f"Test 4 passes")

# Test 5
assert is_valid_test("([)]") == False
print (f"Test 5 passes")

# Test 6
assert is_valid_test("]()]") == False
print (f"Test 6 passes")
```
