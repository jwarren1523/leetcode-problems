```python
# Part 1

def is_palindrome_1(x: int) -> bool:
    if x < 0 : 
        return False 
    if x == 0 : 
        return True
    for i in range(len(str(x))//2):
            if str(x)[i] != str(x)[-1-i]:
                return False
    return True
```

```python
# Part 2

import math

def is_palindrome_2(x: int) -> bool:
    if x < 0 : 
        return False 
    if x == 0 : 
        return True
    order_of_magnitude = math.floor(math.log10(x))
    break_point = 1 + order_of_magnitude // 2
    # print(f"order_of_magnitude: {order_of_magnitude}")
    # print(f"break_point: {break_point}")
    for i in range(break_point):
        first_num = (x // 10**(order_of_magnitude-i)) % 10
        second_num = x // 10**(i) % 10 
        # print("---")
        # print(f"i: {i}")
        # print(f"first_num: {first_num}")
        # print(f"second_num: {second_num}")
        # print("---")
        if first_num != second_num:
            return False
    return True
```

```python
# ChatGPT solution

def is_palindrome_a1(x: int) -> bool:
    if x < 0 : 
        return False 
    s = str(x)
    return s == s[::-1] # Put string in reverse and see if it's the same... basic! 
```

```python
palindrome_test_function = is_palindrome_a1

# Test 1: Valid palindromes
## odd
assert palindrome_test_function(121) is True
assert palindrome_test_function(32423) is True
## even
assert palindrome_test_function(1221) is True
assert palindrome_test_function(324423) is True
print ("Test 1 passes")

# Test 2: negative number is false
assert palindrome_test_function(-5) is False
print ("Test 2 passes")

# Test 3: Not a palindrome and positive
## odd
assert palindrome_test_function(152) is False
assert palindrome_test_function(37423) is False
## even
assert palindrome_test_function(10) is False
assert palindrome_test_function(12345321) is False
print ("Test 3 passes")

# Test 4: zero
assert palindrome_test_function(0) is True
print ("Test 4 passes")

# Test 5: digits
assert palindrome_test_function(1) is True
assert palindrome_test_function(8) is True
print ("Test 5 passes")

# Test 6: Large numbers
assert palindrome_test_function(2147447412) is True   # near 32-bit max
assert palindrome_test_function(2147483647) is False 
print ("Test 6 passes")
```
