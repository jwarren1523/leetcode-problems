```python
def find_substring(s: str, words: list[str]) -> list[int]:
    permutations = get_permutations(words)
    all_indicies = []
    for p in permutations:
        p_indicies = indicies_for_one_permutation(s, p)
        all_indicies.extend(p_indicies)
    list.sort(all_indicies)
    return all_indicies

def get_permutations(words: list[str]) -> dict[int]:
    return permutation_with_first_word("", words)

def permutation_with_first_word(first_word:str, remaining_words:list[str]) -> dict[str]:
    remaining_words_permutations = set()
    if len(remaining_words) == 1:
        return {first_word+remaining_words[0]}
    for i in range(len(remaining_words)):
        popped_list = remaining_words.copy()
        second_word = popped_list.pop(i)
        remaining_words_permutations.update(permutation_with_first_word(second_word, popped_list))
    permutations_with_first_word = set()
    for w in remaining_words_permutations:
        permutations_with_first_word.update({first_word+w})
    return permutations_with_first_word


def indicies_for_one_permutation(s:str, permutation:str) -> list[int]:
    s_copy = s
    length = len(permutation)
    index_count = 0
    indicies = []

    while True:
        idx = s_copy.find(permutation)
        if idx != -1:
            indicies.append(idx+index_count)
            s_copy = s_copy[:idx]+ s_copy[idx+length:]
            index_count += length
        else:
            return indicies

# My sol works if words don't overlap, but if they do, an error happens. Perhaps removing the first letter and all the preceding words of a matched word instead of the whole word would fix this. 
```
